# **Data Analysis AND Data Visualization**

```python
canceled_Percentage = ta['is_canceled'].value_counts(normalize = True)
canceled_Percentage
```

```python
plt.figure(figsize=(6,5))
count = sns.countplot(data = ta, x = "is_canceled", hue = "is_canceled")
count.bar_label(count.containers[0])
count.bar_label(count.containers[1])
plt.legend(["not cancelled", "cancelled"])
plt.title("reservation_status_count", fontsize=20, fontweight='bold', pad=20)
plt.savefig("reservation_status_count.jpg")
plt.show()
```

```python
plt.figure(figsize=(6,5))
count = sns.countplot(data = ta, x = "hotel", hue = "is_canceled", palette = {0:'#f0bc13', 1:'#14addb'})
count.bar_label(count.containers[0])
count.bar_label(count.containers[1])
plt.xlabel('Hotel')
plt.ylabel('Number of Reservation Status')
plt.legend(["not cancelled", "cancelled"])
plt.title("reservation status in Different Hotel", fontsize=20, fontweight='bold', pad=20)
plt.savefig("reservation status in Different Hotel.jpg")
plt.show() 
```

```python
resort_hotel = ta[ta["hotel"]=="Resort Hotel"]
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = ta[ta["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby("reservation_status_date")[['adr']].mean()
resort_hotel
```

```python
city_hotel = city_hotel.groupby("reservation_status_date")[['adr']].mean()
city_hotel
```

```python
plt.figure(figsize=(15,5))
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize=15)
plt.title("Average Daily Rate in City And Resort Hotel", fontsize=20, fontweight='bold', pad=20)
plt.savefig("Average Daily Rate in City And Resort Hotel.jpg")
plt.show()
```

```python
ta['month'] = ta['reservation_status_date'].dt.month
ta['month'] 
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data = ta, x = 'month', hue = 'is_canceled', palette = {0:'#29c4a0', 1:'#944121'})
plt.xlabel('Month')
plt.ylabel('Number of Reservation')
plt.legend(["not cancelled", "cancelled"])
plt.title("Reservation Status Per Month", fontsize=20, fontweight='bold', pad=20)
plt.savefig("Reservation Status Per Month.jpg")
plt.show()
```

```python
plt.figure(figsize=(12,5))
adr_month =  ta[ta['is_canceled']== 1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month',y='adr', data = adr_month,  palette='viridis')
plt.xlabel('Month')
plt.ylabel('Sum of ADR')
plt.title("ADR per Month", fontsize=20, fontweight='bold', pad=20)
plt.savefig("ADR per Month.jpg")
plt.show()
```

```python
cancelled_data = ta[ta['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
top_10_country
```

```python
plt.figure(figsize=(10,10))
color = ['#29c4a0', '#88de18', '#1d4f85', '#72996e','#14addb', '#f0bc13','#944121','pink','#1d4f85', '#fc1a0a']
plt.pie(top_10_country, autopct='%0.2f%%', labels = top_10_country.index, colors = color, startangle = 180, wedgeprops={'linewidth': .25, 'edgecolor': 'white'} )
plt.legend(loc=2)
plt.title("Top 10 country with reservation cancelled", fontsize=20, fontweight='bold', pad=20)
plt.savefig("Top 10 country with reservation cancelled.jpg")
plt.show()
```

```python
cancelled_ta_adr = cancelled_data.groupby('reservation_status_date') [['adr']].mean()
cancelled_ta_adr.reset_index(inplace=True)
cancelled_ta_adr.sort_values('reservation_status_date', inplace=True)
not_cancelled_data = ta[ta['is_canceled']==0]
not_cancelled_ta_adr = not_cancelled_data.groupby('reservation_status_date') [['adr']].mean()
not_cancelled_ta_adr.reset_index(inplace = True)
not_cancelled_ta_adr.sort_values('reservation_status_date', inplace = True)
plt.figure(figsize = (20,6))
plt.plot(not_cancelled_ta_adr['reservation_status_date'], not_cancelled_ta_adr['adr'], label='not cancelled')
plt.plot(cancelled_ta_adr['reservation_status_date'], cancelled_ta_adr['adr'], label='cancelled')
plt.legend()
plt.title('Average Daily Rate',fontsize=30, fontweight='bold', pad=20)
plt.savefig("Average Daily Rate.jpg")
plt.show()
```
